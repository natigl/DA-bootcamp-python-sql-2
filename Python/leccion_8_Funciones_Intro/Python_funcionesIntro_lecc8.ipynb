El objetivo de este ejercicio es convertir un string en un nuevo string en el que cada car치cter del nuevo string es "(" si ese car치cter aparece s칩lo una vez en el string original, o ")" si ese car치cter aparece m치s de una vez en el string. Nuestro c칩digo no tiene que ser case sensitive, es decir, si hay una letra en may칰scula y en min칰scula cuenta como dos apariciones.
Probad la funci칩n para los siguientes strings 

"din"      =>  "((("
"recede"   =>  "()()()"
"Success"  =>  ")())())"
"(( @"     =>  "))((" 
"Ocvl@GamFLAFkixkS" => "((()(()()))))(()("

````python
#las letras repetidas, si aparece mas de una vez tienen que ir en parentesis cerrado y de lo contrario irian en parentesis abierto

def conv_str (palabra):
    palabra = palabra.lower()
    for letra in palabra: 
        if palabra.count(letra) ==1:
            palabra = palabra.replace(letra, '(')
        else:
             palabra = palabra.replace(letra, ')')
    return palabra 
````
