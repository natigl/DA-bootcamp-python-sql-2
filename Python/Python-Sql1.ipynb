- Extraed el números de pedidos que se hayan hecho por dia. Devuelve los resultados en una dataframe.

```python
mybd = mysql.connector.connect(user='root', password='alumnaadalab',
                              host='127.0.0.1',
                              database='Northwind')

mycursor = mybd.cursor()

query = ("""SELECT COUNT(order_id), order_date
            FROM orders
            GROUP BY order_date """)

mycursor.execute(query)
#No imprimimos resultado porque nos da una lista larga, y seleccionamos con el metodo head las 10 primeras filas de la lista de pedidos con sus fechas correspondientes. 
myresult= mycursor.fetchall()

df3 = pd.DataFrame(myresult, columns = ['Pedidos', 'Fecha de Pedido'])

mycursor.close()
mybd.close()
```

- Extraer los pedidos hechos por "Nancy Daviolo". Tendréis que hacer una subquery.

```python
mybd = mysql.connector.connect(user='root', password='alumnaadalab',
                              host='127.0.0.1',
                              database='Northwind')
from mysql.connector import errorcode

mycursor = mybd.cursor()

query = ("""SELECT * FROM orders
            WHERE employee_id IN (
                                SELECT employee_id
                                FROM employees
                                WHERE first_name= 'Nancy' and last_name= 'Davolio')""")

mycursor.execute(query)
#No imprimimos resultado porque nos da una lista larga, y seleccionamos con el metodo head las 10 primeras filas de la lista de pedidos con sus fechas correspondientes. 
myresult= mycursor.fetchall()

df4 = pd.DataFrame (myresult, columns = [ 'order_id', 'customer_id', 'employee_id', 'order_date', 'required_date', 'shipped_date', 'ship_via', 'freight', 'ship_name', 'ship_addrees', 'ship_city', 'ship_region', 'ship_postal_code', 'ship_country'] )
```
